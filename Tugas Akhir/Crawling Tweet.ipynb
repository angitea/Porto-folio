```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
```

```python
# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#BesokSenin kerja since:2020-01-01 until:2022-07-31').get_items()):
    if i>=5000:
        break
    tweets_list2.append([tweet.date, tweet.username, tweet.id, tweet.content])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Date','Username', 'User ID', 'Tweet'])
```
